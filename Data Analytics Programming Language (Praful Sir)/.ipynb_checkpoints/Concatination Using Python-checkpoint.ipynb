# Concatination means static joining tow tables 
# We will use pandas library in python.
# When data gathered in datalake.

```python
# Step 1:- Create two dataframes
import numpy as np
import pandas as pd

# create first Dataframe  "df_1"
df_1 = pd.DataFrame({"Name":["Ram","Shyam","Ham","Mahesh","Dan",np.nan],
                    "Marks":[15,20,15,20,18,17],
                    "Branch":["Mobile","NS","FinTech","BIA","Data","Game"],
                    })

# create first Dataframe  "df_2"
df_2 = pd.DataFrame({"Name":["Virat","Rohit","Ashwin",np.nan,np.nan,"Aayush"],
                    "Marks":[20,22,18,np.nan,22,14],
                    "Branch":["Data","Game","Mobile","NS","FinTech","BIA"],
                    })
```

```python
# Show dataframes
# df_1

df_1.head()
```

```python
# df_2

df_2.head()
```

```python
# Here in df_concat 
```

# concatinate both the Dataframes.



```python
# concatinate both dataframes.

df_concat = pd.concat([df_1,df_2],join = 'inner')
df_concat

```

```python
# here axis = 1 means  it represent's data in horizontal order.
```

```python
df_concat = pd.concat([df_1,df_2],axis = 1)
df_concat
```

```python
df_concat = pd.concat([df_1,df_2],axis = 1, ignore_index = True)
df_concat
```

```python
# Here "ignore_index = true" means after index ends still it continues with next number.
```

```python
df_concat = pd.concat([df_1,df_2],axis = 0, ignore_index = True)
df_concat
```
