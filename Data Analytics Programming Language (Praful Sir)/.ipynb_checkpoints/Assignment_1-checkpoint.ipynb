# # Task 6:- 
## Vertical Concatenation of Customer_1,Customer_2,
## Vertical Concatenation of Employee_1,Employee_2,
## Vertical Concatenation of Order_1,Order_2,
## Vertical Concatenation of Product_1,Product2

```python

# Vertical concatenation of Customer_1 and Customer_2
combined_customer = pd.concat([Customer_1, Customer_2], ignore_index=True)

# Display the combined DataFrame
print("Combined Customer DataFrame:")
print(combined_customer)
```

```python
# Vertical concatenation of Employees_1 and Employees_2
combined_employees = pd.concat([Employee_1, Employee_2], ignore_index=True)

# Display the combined DataFrame
print("Combined Employees DataFrame:")
print(combined_employees)
```

```python
# Vertical concatenation of Orders_1 and Orders_2
combined_orders = pd.concat([Order_1, Order_2], ignore_index=True)

# Display the combined DataFrame
print("Combined Orders DataFrame:")
print(combined_orders)
```

```python
# Vertical concatenation of Product_1 and Product_2
combined_products = pd.concat([Product_1, Product_2], ignore_index=True)

# Display the combined DataFrame
print("Combined Products DataFrame:")
print(combined_products)
```

#  Task 7:-
## Horizontal concatenation (Row wise) on
## (Customer_1 with Customer_2), (Employees_1 - Employees_2), (Orders_1 -
## Orders_2), (Product_1 - Product_2).

```python
# Horizontal concatenation of Customer_1 and Customer_2
combined_customer = pd.concat([Customer_1, Customer_2], axis=1)

# Display the combined DataFrame
print("Combined Customer DataFrame:")
print(combined_customer)
```

```python
# Horizontal concatenation of Employees_1 and Employees_2
combined_employees = pd.concat([Employee_1, Employee_2], axis=1)

# Display the combined DataFrame
print("Combined Employees DataFrame:")
print(combined_employees)
```

```python
# Horizontal concatenation of Orders_1 and Orders_2
combined_orders = pd.concat([Order_1, Order_2], axis=1)

# Display the combined DataFrame
print("Combined Orders DataFrame:")
print(combined_orders)
```

```python
# Horizontal concatenation of Product_1 and Product_2
combined_products = pd.concat([Product_1, Product_2], axis=1)

# Display the combined DataFrame
print("Combined Products DataFrame:")
print(combined_products)
```
