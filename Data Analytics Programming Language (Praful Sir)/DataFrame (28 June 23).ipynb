```python
import pandas as pd
import numpy as np
```

```python
#Make two series.
series1 = pd.Series([1,2,3,4,5],index=['A','B','C','D','E'])
print(series1)
series2=pd.Series([6,7,8,9,10],index=['A','B','C','D','E'])
print(series2)
series3 = pd.Series([11,12,13,14,15],index=['A','B','C','D','E'])
print(series3)
series4 =pd.Series([16,17,18,19,20],index=['A','B','C','D','E'])
print(series4)
series5 =pd.Series([21,22,23,24,25],index=['A','B','C','D','E'])
print(series5)
series6 =pd.Series([26,27,28,29,30],index=['A','B','C','D','E'])
print(series6)
series7 =pd.Series([31,32,33,34,35],index=['A','B','C','D','E'])
print(series7)
series8 =pd.Series([36,37,38,39,40],index=['A','B','C','D','E'])
print(series8)
series9 =pd.Series([41,42,43,44,45],index=['A','B','C','D','E'])
print(series9)
series10 =pd.Series([46,47,48,49,50],index=['A','B','C','D','E'])
print(series10)
series11 =pd.Series([51,52,53,54,55],index=['A','B','C','D','E'])
print(series11)
```

```python
# Now convert those series in 1 D datastructure (Dataframe)
Dataframe_1 = pd.DataFrame({"Column 1":series1,
                           "Column 2":series2,
                           "Column 3":series3,
                           "Column 4":series4,
                           "Column 5":series5,
                           "Column 6":series6,
                           "Column 7":series7,
                           "Column 8":series8,
                           "Column 9":series9,
                           "Column 10":series10,
                           "Column 11":series11})
Dataframe_1
```
